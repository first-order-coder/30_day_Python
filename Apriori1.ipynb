```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header= None)
```

```python
transactions = []
for j in range(0, 7501):
  transactions.append([str(i) for i in df.iloc[j, :]])
```

```python
from apyori import apriori
rules = apriori(transactions= transactions, min_support= 0.003, min_confiedence= 0.2, min_lift= 3, min_length= 2, max_length= 2)
```

```python
results = list(rules)
results
```


## Putting the results well organised into a Pandas DataFrame


```python
def inspect(results):
  lhs = [tuple(result[2][0][0])[0] for result in results]
  rhs = [tuple(result[2][1][0])[0] for result in results]
  support = [result[1] for result in results]
  confidence = [result[2][0][2] for result in results]
  lifts       = [result[2][0][3] for result in results]
  return list(zip(lhs, rhs, support, confidence, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
