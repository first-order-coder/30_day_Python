```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=1)

X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_predict = regressor.predict(X_test)
y_predicted = regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
print(regressor.predict(X_train))
print(y_train)
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, y_train, color='blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

When we use X_test to plot features the density and range of x axis gets smaller so its better to use X_train for X axis.


```python
plt.scatter(X_test, y_test, color='red') #actual values
plt.plot(X_test, regressor.predict(X_test), color='blue') #predicted value of X_test
plt.show()
```

```python
plt.scatter(X_test, y_test, color='red') #actual values
plt.plot(X_train, regressor.predict(X_train), color='blue') #predicted value of X_test
plt.show()
```

regressor.predict(X_train) will predict the new target variable according to found coefficents and intercepts of the regression line function. but there is very small difference between actual y values and predicted ones. But when we plot these two things regressor.predict(X_train) and y_train we get whole two different outputs of plots.
